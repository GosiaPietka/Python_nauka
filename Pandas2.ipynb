```python
import sqlite3
conn = sqlite3.connect('pandas_exercise_01')

dfs = {}

for name in names:
    df = pd.read_csv(f"Exercise01/{name}.csv", comment='#')
    dfs[name] = df
    df.to_sql(name, con=conn, if_exists='replace')
```
