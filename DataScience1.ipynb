```python
review_points_mean = reviews_df.points.mean()

def remean(points):
    return points - review_points_mean

reviews_df.points.map(remean)
```

```python
reviews_df.points.map(lambda points: points - review_points_mean) #funkcja istnieje tylko w ramach tej komendy
```

```python
# apply

def remean_points(row):
    row.points = row.points - review_points_mean
    return row

reviews_df.apply(remean_points, axis='columns')
```
