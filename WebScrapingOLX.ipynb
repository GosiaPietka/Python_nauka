```python
import requests
import bs4
from bs4 import BeautifulSoup
import pandas as pd
```

```python
page = requests.get("https://www.olx.pl/d/oferty/q-owczarek-niemiecki/")

soup = BeautifulSoup(page.content, 'html.parser')
```

```python
offers = soup.find_all('div', class_='css-qfzx1y')
```

```python
offers
```

```python
titles = [offer.find('h6').get_text() for offer in offers]
titles
```

```python
prices = [offer.find('p', class_='css-wpfvmn-Text').get_text() 
 for offer in offers]
```

```python
prices
```

```python
links = [offer.parent['href'] for offer in offers]
```

```python
olx_home = "https://www.olx.pl"
links = [f"{olx_home}{link}" for link in links]
```

```python
links
```

```python
pd.DataFrame({
    'title': titles,
    'price': prices,
    'link': links,
})
```

```python
offers_df = pd.DataFrame({
    'title': titles,
    'price': prices,
    'link': links,
})
```

```python
offers_df.to_excel('owczarki.xlsx') #generowanie dok exel
```
